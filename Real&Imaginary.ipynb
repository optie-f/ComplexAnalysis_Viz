# 複素関数の実部と虚部

```julia
using Plots
using Interact
import ProgressMeter
using LaTeXStrings
using IJulia

function displayfile(filename)
    display("text/html", """<img src="$filename">""")
end

gr(
    titlefont=Plots.font("sans-serif", 12),
    legendfont=Plots.font("sans-serif", 8),
    guidefont=Plots.font("sans-serif", 10),
    tickfont=Plots.font("sans-serif", 8),
    markersize=4, markerstrokewidth = 1, 
    markercolor= "#FF445B",markerstrokecolor="#4A4A4A",
    linewidth=2
)

ENV["PLOTS_TEST"] = "true"  #出力画像サイズ6倍問題回避

dir = "gif/"
fps = 30
```

```julia
function plot2(f;xrange=10,yrange=10,horizontal=true)
    axiscolor = "#494850"

    x = linspace(-xrange,xrange,50)
    y = linspace(-yrange,yrange,50)
    z = [i + j.*im for i in x, j in y]'

    X = real.(f(z))
    Y = imag.(f(z))

    p1 = plot(title="real", aspect_ratio=1)
    p2 = plot(title="imag",aspect_ratio=1)

    plot!(p1, x, y, X, st=:contourf)
    plot!(p2, x, y, Y, st=:contourf)

    vline!(p1,[0],line=(axiscolor, 1))
    hline!(p1,[0],line=(axiscolor, 1))
    vline!(p2,[0],line=(axiscolor, 1))
    hline!(p2,[0],line=(axiscolor, 1))

    θ = linspace(0,2π,60)
    plot!(p1, sin.(θ), cos.(θ), line=(axiscolor, 1))
    plot!(p2, sin.(θ), cos.(θ), line=(axiscolor, 1))
    
    if horizontal
        plot(p1,p2,layout=(1,2))
    else 
        plot(p1,p2,layout=(2,1))
    end
end
```

## 正則な関数

```julia
function f1(z)
    log.(z)
end

plot2(f1)
```

```julia
function f2(z)
    sin.(z)
end

plot2(f2,xrange=10,yrange=1,horizontal=false)
```

```julia
function f3(z)
    sinh.(z)
end

plot2(f3,xrange=5,yrange=10)
```

```julia
function f4(z)
    tan.(z)
end

plot2(f4,xrange=2π,yrange=2π)
```

```julia
function f4(z)
    exp.(z)
end

plot2(f4)
```

```julia
function f5(z)
    z.^2
end

plot2(f5)
```

```julia
function f5(z)
    z.^(3)
end

plot2(f5)
```

```julia
function f7(z)
    z.^(-1)
end

plot2(f7,xrange=2,yrange=2)
```

```julia
function f7(z)
    α= 0.5+0.5im
    (z-α).^(-1)
end

plot2(f7,xrange=1,yrange=1)
```

```julia
function f7(z)
    α = 0.5
    β = -0.5
    ((z-α).*(z-β)).^(-1)
end

plot2(f7,xrange=2,yrange=2)
```

```julia
function f9(z)
    sin.(z.^(-1)).^(-1)
end

plot2(f9,xrange=1,yrange=1)
```

```julia
function f9(z)
    sin.(z) ./ z
end

plot2(f9,xrange=12,yrange=4,horizontal=false)
```

## 正則でない関数

```julia
function f6(z)
    abs.(z)
end

plot2(f6)
```

```julia
function f7(z)
    conj.(z)
end

plot2(f7)
```
